Dataframes

- Dataframes are the main data structure used in pandas, and its basically a table

```python
data = [
    ['Nissan', 'Stanza', 1991, 138, 4, 'MANUAL', 'sedan', 2000],
    ['Hyundai', 'Sonata', 2017, None, 4, 'AUTOMATIC', 'Sedan', 27150],
    ['Lotus', 'Elise', 2010, 218, 4, 'MANUAL', 'convertible', 54990],
    ['GMC', 'Acadia',  2017, 194, 4, 'AUTOMATIC', '4dr SUV', 34450],
    ['Nissan', 'Frontier', 2017, 261, 6, 'MANUAL', 'Pickup', 32340],
]

columns = [
    'Make', 'Model', 'Year', 'Engine HP', 'Engine Cylinders',
    'Transmission Type', 'Vehicle_Style', 'MSRP'
]
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(columns)
```

```python
df = pd.DataFrame(data, columns=columns)
df
```

- A table consists of many Series

```python
df.Make
df['Make']
df['Engine HP']     # for column heading with space, we use the brackets
```

```python
df[['Make', 'Model', 'MSRP']]       # A way of accessing multiple columns
```

```python
# To add a column to the dataframe
df['id'] = [1,2,3,4,5]
df

# We can also change the numbering of the columns
df['id'] = [5,10,15,20,25]
df
```

Element-wise Operations

```python
# We can carry out the same operations as in numpy
df['Engine HP'] * 2
df['Engine HP'] / 100
```

```python
# Using Logical Operators


df['Year'] >= 2013
```

Filtering

```python
# To filter a particular attribute of the table
df[
    df['Make'] == 'Nissan'
]

df[
    df['Year'] >= 2015
]
```

```python
# We can combine filters too

df[
    (df['Make'] == 'Nissan') & (df['Year'] >= 2015)
]
```

String Operators

```python
# We can replace strings in a table
df
```

```python
# We can replace strings using the "replace" keyword
'machine learning zoomcamp'.replace(' ', '_')
```

```python
# we can convert a particular column to lower case
df['Vehicle_Style'].str.lower()
```

```python
# we can also combine operations
df['Vehicle_Style'] = df['Vehicle_Style'].str.replace(' ', '_').str.lower()
df
# This replaces the empty spaces with "_" and converts to lower case for the vehicle style column
```

Summarizing Operations

```python
df.MSRP
df.MSRP.max()
df.MSRP.min()
df.MSRP.median()
df.MSRP.mean()
```

```python
# "describe gives us all the useful statistics"
df.MSRP.describe()
```

```python
# we can also describe the entire dataframe
# it will describe all the numeric columns
df.describe()
```

```python
# we can check for unique values in a column
# this is a string operation where we check the number of unique values are in a column or in a table
df['Transmission Type'].nunique()
```

```python
df.nunique()
```

Missing Values (Nan)

```python
# to know if a value is missing or not, we use "isnull()"
df.isnull()
```

```python
# This tells the number of missing values in each column
df.isnull().sum()
```

Grouping

```python
# we want to group by transmission type, and then find the average price for each group
'''
    SELECT 
    transmission_type,
    AVG(MSRP)
FROM
    cars
GROUP BY
    transmission_type
'''
```

```python
df.groupby('Transmission Type').MSRP.mean()
df.groupby('Transmission Type').MSRP.min()
df.groupby('Transmission Type').MSRP.max()
```
